```python
Target_1 = Mixture([Normal(-6, 1), Normal(-3, 1), Normal(0, 1)], 
                                    [1, 2, 4])
```

```python
samplers = [
RepeatedSampler({"p":Target_1, "q": RandomWalkGaussianProposal(2)}),
RepeatedSampler({"p":Target_1, "q": MALA(1, Target_1.score)}),
RepeatedSampler({"p":Target_1, "q": MALA(1, Target_1.score, str_rep="ULA")}, sampler=ULA),
            ]


RepeatedSamplerWrapper().plot(samplers, sample_args={"repetitions": 500, "sample_length": 200, "x0": generate_x0, "burnin": 0}, 
                            plot_args={"window_width": 20})
```
