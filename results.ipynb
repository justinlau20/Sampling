```python
from infrastructure import Proposal, MetropolisHastings, Normal, RandomWalkGaussianProposal, AggregatePlotter, IndependentGaussianProposal, Mixture
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(np.linspace(-3, 3),  Mixture(Normal(-1, 1), Normal(1, 1), 0.7).log_prob(torch.tensor(np.linspace(-3, 3))))
```

```python
def _plot_RandomWalk():
    MH = MetropolisHastings(Mixture(Normal(-1, 1), Normal(1, 1), 0.7), RandomWalkGaussianProposal(1))
    MH.sample(300, torch.tensor(0))
    MH.plot(window_width=50);
_plot_RandomWalk()
```

```python
def _plot_RandomWalk_aggregate():
    ap = AggregatePlotter()
    MH = MetropolisHastings(Mixture(Normal(-1, 1), Normal(1, 1), 0.7), RandomWalkGaussianProposal(1))
    ap.sample(300, 200, MH, x0=1, window_width=10)
    ap.plot();
_plot_RandomWalk_aggregate()
```

```python
def _plot_RandomWalk():
    MH = MetropolisHastings(Normal(0, 1), RandomWalkGaussianProposal(1))
    MH.sample(300, torch.tensor(5))
    MH.plot(window_width=50);
_plot_RandomWalk()
```
